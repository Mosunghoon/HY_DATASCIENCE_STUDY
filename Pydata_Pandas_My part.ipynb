#### 5.5.1 계층 순서 바꾸고 정렬하기 

swaplevel 메서드를 활용해서 사용가능

```python
frame.swaplevel('key1','key2') # 이름 순서를 바꿀 수 있다.
```

```python
frame
```

```python
frame.sortlevel(1) # sortlevel은 단일 계층에 속한 데이터를 정렬함.
```

```python
frame.swaplevel(0,1).sortlevel(1) # 이렇게 복합적으로 해서 인덱스를 옮겨서 작성할 수 있다.
```

```python
dframe = pd.DataFrame(np.arange(18).reshape((6,3)),
                  index = [['a','a','a','b','b','b'],[1,2,3,1,2,3]],
                  columns = [['Ohio','Ohio', 'Colorado'],
                            ['Green','Red','Green']])
```

```python
dframe
```

```python
dframe.swaplevel(0,1).sortlevel(0)
```

#### 5.5.2 단계별 요약통계 

```python
frame.sum(level='key2') 
```

```python
frame.sum(level='color', axis = 1) # 이번에는 컬러라는 컬럼 인덱스를 통해 합산함.
```

```python
frame.sum(level='state',axis = 1) # 주 이름이라는 컬럼 인덱스를 통해 합산함.
```

#### 5.5.3 DataFrame의 칼럼 사용하기 

```python
sframe = DataFrame({'a': range(7), 'b': range(7,0,-1), 'c': ['one','one','one','two','two','two','two'],'d':[0,1,2,0,1,2,3]})
```

```python
sframe
```

```python
sframe2 = sframe.set_index(['c','d']) #c와 d는 명목형 변수 컬럼으로, 이처럼 set_index를 통해 컬럼을 인덱스로 사용할 수 있다.
```

```python
sframe2
```

```python
sframe.set_index(['c','d'],drop=False) #drop=False는 칼럼의 레코드값을 명시적으로 남겨놓으라는 인자이다.
```

```python
sframe2.reset_index() #reset_index()를 사용하게 되면 본래 인덱스로 활용되던 c와 d의 데이터가 다시 레코드값으로 돌아오게 된다.
```
