```python
for i in range(10):
    try:
        result = 10/i
    except ZeroDivisionError:
        print("Not divded by 0")
    else:
        print(10/i)
```

```python
for i in range(10):
    try:
        result = 10//i
    except ZeroDivisionError:
        print("Not divded by 0")
    finally:
        print("종료")
```

```python
while True:
    value = input("변환할 정수 값을 입력해주세요")
    for digit in value:
        if digit not in "0123456789":
            raise ValueError('숫자값을 입력하지 않으셨습니다')
    print("정수값을 변환된 숫자 -", int(value))
```
