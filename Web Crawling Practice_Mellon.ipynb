```python
## 리퀘스트. 사람처럼 웹 정보 요청.
## CSS Selector 중요하다. HTML을 예쁘게 꾸미는 게 CSS Selector
## 부트스트랩
## tag name / tag id / tag class names

## CSS Selector Syntax
## 일반적으로
## HTML이 있고 그 밑에 head, body, div 이런 순서대로 계층이 구성되어 있다.
## 그 안에 <table class = 뭐시기뭐시기> 이런 형식으로 만들어진다.
## 어떤 경우에는 id만 있는 경우도 있다.
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://www.melon.com/chart/index.htm"

html = requests.get(url).text

soup = BeautifulSoup(html, 'html.parser')

tag_list = soup.select('#chartListObj tr .wrap_song_info a[href*=playSong]')
for idx, tag in enumerate(tag_list, 1):
    print(idx, tag.text)

    
## idx는 프로그래밍 할 때 순서를 정의해준다
## enumerate 뭔가 정보가 들어올 때 1 값을 위로 올려주는 기능을 한다.
```
